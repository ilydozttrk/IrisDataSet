```python
#kütüphaneyi içe aktarma
import sklearn, pandas as pd,numpy as np #import= bir kütüphaneyi projeye dahil eder.
from sklearn.datasets import load_iris #load_iris= hazir iris veri setini yükleyen fonksiyon.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #LogisticRegression = basit bir maakine öğrenmesi modeli.
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report 
# accuracy_score, confusion_matrix, classification_report= sonuçları değerlendirmeye yarayan ölçüm araçlarıdır.
```

```python
#veri yükleme
iris = load_iris(as_frame=True) # veriyi tablo (DataFrame) getirir
x = iris.data #girdiler, yani her çiçeğin sayısal özellikleri (ör: taç yaprağı uzunluğu)
y = iris. target #çıkış (etiket): hangi çiçek türü

display(x.head()) #ilk 5 satırı gösterir
display(y.head()) 

print("özellik adları:", x.columns.tolist()) #sütun isimlerini listeler
print("sınıf isimleri:", iris.target_names.tolist()) #türlerin isimleri
```

```python
#eksik veri ve sınıf dengesi kontrolü
null_counts = x.isnull().sum() #her sütunda eksik veri var mı, sayar
print("eksik deper sayıları:\n", null_counts)

print("\nSınıf dağılımı (y):")
print(y.value_counts().sort_index(), "\n") #her sınıftan kaç örnek olduğunu gösterir.
#bu adım, modelin sağlıklı öğrenebilmesi için veri kalitesini kontrol eder.
```

```python
#veriyi bölme
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y)
#%80 eğitim, %20 test olarak ayırır
#random_state=42 rastgeleliği sabit tutar (aynı sonucu tekrar alırsın)
#stratify=y ise sınıfları orantılı olarak dağıtır 
print(x_train.shape, x_test.shape) #boyutları kontrol ederiz (örneğin 120 eğitim, 30 test örneği
```

```python
#model oluşturma ve eğitme
model =LogisticRegression(max_iter=1000) # LogisticRegression()=doğru sınıfa ait olasılıkları hesaplayan bir model
#max_iter=1000 ise öğrenme sırasında iterasyon sınırını attırır (yoksa bazen uyarı:yakınsanmadı hatası alırız)
model.fit(x_train, y_train) #modeli eğitir, yani veriyi öğretir
```

```python
#modeli değerlendirme
train_acc= model.score(x_train, y_train) #score doğruluk oranını döndürür (doğru tahmin sayısı/ toplam)
test_acc= model.score(x_test, y_test)

print(f"eğitim doğruluğu: {train_acc:.3f}") #sayıyı 3 ondalıklı yazdırır
print(f"test doğruluğu: {test_acc:.3f}")

#bu iki değeri karşılaştırırsak:
#eğitim çok yüksek, test düşükse = overfitting (aşırı öğrenme)
#ikisi de düşükse= model yeterince öğrenememiştir

```

```python
#karışıklık matrisi ve sınıflandırma raporu
y_pred= model.predict(x_test) #.predict(X_test)= modelin test verisinde ne tahmin ettiğini döndürür
print("karışıklık matrisi:")
print(confusion_matrix(y_test, y_pred)) #gerçek ve tahmin edilen sınıfları karşılaştırır. köşegen üzerindekiler doğru tahminlerdir.
print("\nSınıflandırma raporu:")
print(classification_report(y_test, y_pred, target_names= iris.target_names)) 
#classification_report= her sınıf için:
#precision(kesinlik)= model 'setosa' dediyse, gerçekten o mu?
#recall(duyarlılık)= tüm gerçek setosaların kaçı bulundu
#f1-score= precision + recall ortalaması
```
